```python
!pip install pulp==2.7.0

import pulp

# Define the problem
prob = pulp.LpProblem("TransportationProblem", pulp.LpMinimize)

# Define the decision variables
a = pulp.LpVariable("a", lowBound=0)
b = pulp.LpVariable("b", lowBound=0)
c = pulp.LpVariable("c", lowBound=0)
d = pulp.LpVariable("d", lowBound=0)

# Define the objective function
prob += 4*a + 6*b + 5*c + 3*d, "Total Transportation Cost"

# Define the constraints
prob += a + b <= 100, "Constraint 1"
prob += c + d <= 120, "Constraint 2"
prob += a + c == 80, "Constraint 3"
prob += b + d == 140, "Constraint 4"

# Solve the problem
prob.solve()

# Print the results
print("Status:", pulp.LpStatus[prob.status])
print("Optimal Solution:")
print("a =", a.varValue)
print("b =", b.varValue)
print("c =", c.varValue)
print("d =", d.varValue)
print("Total Transportation Cost =", pulp.value(prob.objective))
```
